```python
import pandas as pd
import numpy as np
```

```python
file_path = "Data/Animal permits - 1 July 2020.csv"
dog_list = pd.read_csv(file_path)
dog_list
```

```python
# loop through each column, check if there are null values, 
#sum them up, and print out a readable total:
for column in dog_list.columns:
    print(f"Column {column} has {dog_list[column].isnull().sum()} null values")
```

```python
#data type
dog_list.dtypes
```

```python
#check Permit Groupoptions
dog_list.groupby("Permit Group").count().index
```

```python
#check Permit Name options
dog_list.groupby("Permit Name").count().index
```

```python
#check Permit Status options
dog_list.groupby("Permit Status").count().index
```

```python
#Remove all Permit Namethat are not DOG
dog_list_2= dog_list.loc[dog_list['Permit Name'].str.contains('Dog')]
dog_list_2
```

```python
#check Permit Name options
dog_list_2.groupby("Permit Name").count().index
```

```python
#count breed value
dog_list_2['Animal: Breed'].value_counts(dropna=False)
```
