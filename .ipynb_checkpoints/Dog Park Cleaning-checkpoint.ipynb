```python
#create a crosstab as a pivot table to count how many breed
pivot_dog=pd.crosstab(dog_list_3.Suburb, dog_list_3.Breed)
pivot_dog
```

```python
# Rename the columns from SUBURB TO Suburb
park_list_2 = park_list.rename(columns = {"SUBURB": "Suburb"}, inplace = False)

park_list_2
```
